```python
import pandas as pd

df=pd.read_excel('/content/purchase-list.xlsx')
df.head()
```

```python
df=df.transpose()
df=df.iloc[1:,:]
df.reset_index(inplace=True)
df.fillna(0, inplace=True)
df.head()
```

```python
df.rename(columns={'index': 'month'}, inplace=True)
df
```

```python
df['month'] = df['month'].str.split().str[1]
df['month'] = df['month'].str[:-1]
df.head()
```

```python
df['month'] = df['month'].str.extract(r'(\b\w+\b)', expand=False)
df.head()
```

```python
df['Total_Sum'] = df.iloc[:,1:].sum(axis=1)
print(df[['month', 'Total_Sum']])
```

```python
new_df=df.groupby('month')['Total_Sum'].sum().reset_index()
new_df
```

```python
new_df.to_excel('SalesInMonths.xlsx', index=False)
```
